### 1. In how many cities Summer Olympics is held so far?

```python
import pandas as pd
df = pd.read_csv("summer.csv")
city=list()
for i in range(len(df)):
    if df['City'][i] not in city:
        city.append(df['City'][i])
print("The no of cities in which it is held is",len(city))
```

### 2. Which sport is having most number of Gold Medals so far? (Top 5)

```python
import pandas as pd
df = pd.read_csv("summer.csv")
sport={}
for i in range(len(df)):
    if df['Sport'][i] not in sport.keys():
        sport[df['Sport'][i]]=0
        if df['Medal'][i]=='Gold':
            sport[df['Sport'][i]]=1
    else:
        if df['Medal'][i]=='Gold':
            sport[df['Sport'][i]]+=1
i=1
while(i<=5):
    mx=max(sport.values())
    for ky, val in sport.items():
        if val==mx:
            print(i," Sport: ", ky, " Gold: ",val)
            del sport[ky]
            break
    i+=1
```

### 6. In which year India won first Gold Medal in Summer Olympics?

```python
import pandas as pd
df = pd.read_csv("summer.csv")
yr=list()
for i in range(len(df)):
    if df['Country'][i] == 'IND':
        if df['Medal'][i] =='Gold':
            yr.append(df['Year'][i])
if len(yr)==0:
    print("India havent won a Gold medal yet")
else:
    print("India won its first medal in",min(yr))
```

### 8. Which sport is having most female Gold Medalists? (Top 5)

```python
import pandas as pd
df = pd.read_csv("summer.csv")
sport={}
for i in range(len(df)):
    if df['Sport'][i] not in sport.keys():
        sport[df['Sport'][i]]=0
        if df['Gender'][i]=='Women':
            if df['Medal'][i]=='Gold':
                sport[df['Sport'][i]]=1
    else:
        if df['Gender'][i]=='Women':
            if df['Medal'][i]=='Gold':
                sport[df['Sport'][i]]+=1
i=1
while(i<=5):
    mx=max(sport.values())
    for ky, val in sport.items():
        if val==mx:
            print(i," Sport: ", ky, " Gold: ",val)
            del sport[ky]
            break
    i+=1
```
